```python
from nltk.corpus import stopwords
```

```python
Except = ['no']
def remove_stopwords(text):
  new_text = []
  for word in text.split():
    if word in stopwords.words('english') and word not in Except:
      new_text.append('')
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
new_text = remove_stopwords(new_text)
```
