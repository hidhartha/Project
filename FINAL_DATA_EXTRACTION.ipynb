```python
import pandas as pd
df = pd.DataFrame()
```

```python
for i in range(len(file_names)):
  with open(file_names[i]) as f:
    x= f.read()

  doc = mp(x)
  parties = ''
  date = ''
  court = ''
  author = ''
  for ent in doc.ents:
    if ent.label_ == 'PARTIES':
      parties = parties + ent.text
    elif ent.label_ == 'DATE OF JUDGEMENT':
  
      date = date + ent.text
    elif ent.label_ == 'AUTHOR':
  
      author = author + ent.text
    elif ent.label_ == 'COURT OF JUDGEMENT':
  
      court = court + ent.text  
    else:
      continue
  lst = []
  lst.append(parties)
  lst.append(date)
  lst.append(court)
  lst.append(author)
  dict = { 'PARTIES': lst[0], "DATE": lst[1], "COURT": lst[2], "AUTHOR": lst[3]}
  df = df.append(dict, ignore_index=True)
  
```

```python
df['Court_Name'] = df['COURT']+ df['Court'].astype(str)
```

```python
df = df.drop(['COURT', 'Court', 'DATE'], axis=1)
```
